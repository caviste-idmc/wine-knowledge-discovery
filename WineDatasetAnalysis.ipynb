# Paspailleur analysis

```python
df = pd.read_csv("../data/preprocessed/WineDatasetPAS.csv")
df.head()
```

```python
df.drop(columns=["Price_bin", "ABV_bin", "Vintage_bin", "Title"], inplace=True)
```

```python
df.info()
```

```python
CharacteristicsPattern = psp.pattern_factory(psp.bip.ItemSetPattern)
StylePattern = psp.pattern_factory(psp.bip.ItemSetPattern)

PricePattern = psp.pattern_factory(
    psp.bip.ClosedIntervalPattern,
    BoundsUniverse=(
        0, 15, 25, 50, 1000
    )
)
ABVPattern = psp.pattern_factory(
    psp.bip.ClosedIntervalPattern,
    BoundsUniverse=(
        0, 12.5, 13.5, 14.5, 50
    )
)
VintagePattern = psp.pattern_factory(
    psp.bip.ClosedIntervalPattern,
    BoundsUniverse=(
        0, 2015, 2020, 2030
    )
)


GrapePattern = psp.pattern_factory(psp.bip.CategorySetPattern, Universe=set(df["Grape"].dropna().unique()))
ClosurePattern = psp.pattern_factory(psp.bip.CategorySetPattern, Universe=set(df["Closure"].dropna().unique()))
CountryPattern = psp.pattern_factory(psp.bip.CategorySetPattern, Universe=set(df["Country"].dropna().unique()))
TypePattern = psp.pattern_factory(psp.bip.CategorySetPattern, Universe=set(df["Type"].dropna().unique()))
RegionPattern = psp.pattern_factory(psp.bip.CategorySetPattern, Universe=set(df["Region"].dropna().unique()))
```

```python
class WinePattern(psp.bip.CartesianPattern):
    DimensionTypes = {
        'Price': PricePattern,
        'Grape': GrapePattern,
        'Closure': ClosurePattern,
        'Country': CountryPattern,
        'Characteristics': CharacteristicsPattern,
        'Type': TypePattern,
        'ABV': ABVPattern,
        'Region': RegionPattern,
        'Style': StylePattern,
        'Vintage': VintagePattern,
    }
```

```python
df = df[['Price', 'Grape', 'Closure', 'Country', 'Characteristics', 'Type', 'ABV', 'Region', 'Style', 'Vintage']]
```

```python
for f in df.columns:
    if df[f].isna().any():
        p = WinePattern.DimensionTypes[f].get_min_pattern()
        df[f] = df[f].fillna(p)
        print(f"Found NaN values in dimension {f}. Replace them by pattern: {p}")
```

```python
ps = psp.PatternStructure(WinePattern)
ps.fit(df.to_dict('index'), min_atom_support=0.1, use_tqdm=True)
print('#objects x #atomic_patterns', ps.shape)
```

```python
concepts = ps.mine_concepts(min_delta_stability=30, min_support=500, algorithm='gSofia', use_tqdm=True)
```

```python
print(f"# concepts: {len(concepts):,}")
extent, intent = concepts[50]
print("Concept #50")
print(f"* objects in the concept: {list(extent)[:10]} (some of them)")
print(f"* pattern of the concept: {intent}")
```

```python
implications = ps.mine_implications(
    min_delta_stability=30, min_support=500, max_key_length=4,
    algorithm='gSofia',
    reduce_conclusions=True, use_tqdm=True
)
print(len(implications))
```

## Subgroups

### Cheap white wine

```python
df = pd.read_csv("../data/preprocessed/WineDatasetPAS.csv")
df.drop(columns=["Price_bin", "ABV_bin", "Vintage_bin", "Title"], inplace=True)
df = df[['Price', 'Grape', 'Closure', 'Country', 'Characteristics', 'Type', 'ABV', 'Region', 'Style', 'Vintage']]
```

```python
df.head()
```

```python
goal_objects = set(
    df[(df['Type'] == 'Red') & (df['Characteristics'].apply(lambda s: 'Red Cherry' in s or 'Cedar' in s))].index
)
subgroups_iterator = ps.iter_subgroups(
    goal_objects=goal_objects,
    quality_measure='Precision', quality_threshold=0.65,
    max_length=2,
    use_tqdm=True
)
```

```python
subgroups = list(subgroups_iterator)
print(len(subgroups))
```

```python
subgroups
```

```python
# Order subgroups 1) by simplicity of pattern, 2) by their quality
subgroups = sorted(subgroups, key=lambda sg_data: (len(sg_data.pattern), -sg_data.quality_value))
print("Subgroups of Red Whine with either Red Cherry or Cedar Characteristic:")
for pattern, objects, quality, quality_name in subgroups:
    print(f"Pattern: {pattern}")
    print(f"{quality_name}: {quality:.2%}, Support: {len(objects)} ({ps.measure_frequency(pattern):.0%} of data)")
    print()
```

```python
df.columns
```

```python
df = df[['Grape', 'Type', 'Region', 'Style']]
```

```python
class WinePattern(psp.bip.CartesianPattern):
    DimensionTypes = {
        'Grape': GrapePattern,
        # 'Country': CountryPattern,
        'Type': TypePattern,
        # 'ABV': ABVPattern,
        'Region': RegionPattern,
        'Style': StylePattern,
    }
```

```python
for f in df.columns:
    if df[f].isna().any():
        p = WinePattern.DimensionTypes[f].get_min_pattern()
        df[f] = df[f].fillna(p)
        print(f"Found NaN values in dimension {f}. Replace them by pattern: {p}")
```

```python
ps = psp.PatternStructure(WinePattern)
ps.fit(df.to_dict('index'), min_atom_support=0.1, use_tqdm=True)
print('#objects x #atomic_patterns', ps.shape)
```

```python
concepts = ps.mine_concepts(min_delta_stability=30, min_support=500, algorithm='gSofia', use_tqdm=True)
len(concepts)
```

```python
patterns = [
    WinePattern.get_min_pattern(),
]

for extent, intent in concepts[:10]:
    patterns.extend([
        WinePattern(intent),
    ])

len(patterns)
```

```python
node_labels = [str(p)+'<br>'+f"support: {ps.measure_support(p):.0f}" for p in patterns]
direct_subpatterns = [{patterns.index(q) for q in patterns if q > p and not any(q > r > p for r in patterns)} for p in patterns]

print(csp.io.to_mermaid_diagram(node_labels, direct_subpatterns))
```

```python
euler_df = dict()
for p in patterns:
  extent = ps.extent(p)
  euler_df[str(p)] = [g in extent for g in ps.objects]

euler_df = pd.DataFrame(euler_df)
euler_df
```
